```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('ASN113O_SOL31081H_rep2_100ns_chainB.dat', sep='\t', header=None, names=['Frame', 'Distancia_1'])
df2 = pd.read_csv('SEROG_HISNE_rep2_100ns_chainB.dat', sep='\t', header=None, names=['Frame', 'Distancia_2'])
df_combined = pd.merge(df1, df2, on='Frame')
#correlation coefficient between the columns 
correlation = df_combined['Distancia_1'].corr(df_combined['Distancia_2'])
#output em floiningpoint 
df_filtered = df_combined.loc[
    (df_combined['Distancia_1'] <= 3 ) & 
    (df_combined['Distancia_2'] <= 3 )
]
correlation_filtered = df_filtered['Distancia_1'].corr(df_filtered['Distancia_2'])
```

```python
correlation_matrix = df_filtered.corr()
plt.figure(figsize=(12,8))
sns.heatmap(correlation_matrix, annot=True, cmap='rainbow', fmt=".2f")
plt.title("Correlation matrix")
```

```python
plt.acorr(correlation_filtered, maxlags=100)
plt.show()
```

```python
plt.figure(figsize=(26, 14))
plt.scatter(df_combined['Distancia_1'], df_combined['Distancia_2'], alpha=0.7, marker='*', c='pink', s=500)
plt.title(f'Correlation: {correlation:.2f}', fontsize=40)
plt.xlabel('d1 - x', fontsize=40, c='red')
plt.ylabel('d2 - y', fontsize=40)
plt.grid(True)
plt.savefig('correlation.png')
```
